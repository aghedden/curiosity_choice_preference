```python
import random
import string
```

```python
# make test (TST) passwords

def generate_passwords(num_passwords):
   passwords = []
   characters = string.ascii_uppercase + string.digits
   
   for i in range(num_passwords):
       random_part = ''.join(random.choice(characters) for _ in range(6))
       sequence_num = str(i + 1).zfill(3)
       password = f"{random_part}TST{sequence_num}"
       passwords.append(password)
   
   return passwords

num_passwords = 10
generated_passwords = generate_passwords(num_passwords)

for password in generated_passwords:
   print(password)
```
